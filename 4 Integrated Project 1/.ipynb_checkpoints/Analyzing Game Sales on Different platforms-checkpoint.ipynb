```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats as st
warnings.filterwarnings("ignore")
games = pd.read_csv('/datasets/games.csv')
print(games.info())
games.head()
```

### Conclusion

On this step we have opened data using `read_csv()` method and displayed it via `head()` method. And investigated more information `info()` method. Some columns have missing elements and some columns have an different datatypes

### Step 2. Preprocessing data

```python
games.columns = ['name','platform','year_of_release','genre','na_sales','eu_sales','jp_sales','other_sales','critic_score','user_score','rating']
games['user_score'] = games['user_score'].replace('tbd',None)
games = games.convert_dtypes()
print(games.info())
games['user_score'] = pd.to_numeric(games['user_score'],errors='coerce')
games.info()
```

Here we have replaced column names with lowercase letters. And we have used `convert_dtypes()` method, which is converts columns in most cases to optimal datatype. `replace()` method to handling `tbd` values on `user_score` column. We have changed them into `None` values. By checking dataframe we have seen that `user_score` not converted rightly. So, we have used `to_numeric()` method 

```python
print('The number of duplicated rows in dataframe:',games.duplicated().sum())

games.dropna(subset=['name','genre'],inplace=True)

games['rating'] = games['rating'].fillna("NOT GIVEN")
games['critic_score'] = games['critic_score'].fillna(games.groupby(['genre'])['critic_score'].transform('median'))
games['user_score'] = games['user_score'].fillna(games.groupby(['genre'])['user_score'].transform('median'))
games.info()
```

Here we have checked for duplicated rows via `duplicated()` method. And we have dropped missimg values of `name` and `genre` column(only 2 rows affecting on data will be very very small which we will ignore). Then missing values of the `rating` column we filled with `"NOT GIVEN"`. And we filled other columns by connections on the data.  `user_score`,`critic_score` have connection on `genres`. We have filled them `groupby` columns which have a connection and found their median to keep data more clear(To keeping data affecting of outlier).And we did not do nothing with `platform` column, because there are only 272 missing values(about 1.6%).

```python
def calc_total_sales(row):
    eu_sales = row['eu_sales']
    na_sales = row['na_sales']
    jp_sales = row['jp_sales']
    other_sales = row['other_sales']
    return (eu_sales + na_sales + jp_sales + other_sales)
games['total_sales'] = games.apply(calc_total_sales,axis=1)
games.head()
```

### Conclusion

On this step we have handled missing values and convert datatypes to needed ones, calculated `total_sales`.COnvertion datatypes makes us easier work with the data later. We converted `critic_score` into `int`,because there is no floating point numbers after 0, it is more convenient working with `int` type rather than `float`. And we converted `user_score` to `float` from `object`. The reason of the filling missing values it is more convenient working with full data. And we have filled missing values with the `median` for avoiding affecting the data.

### Step5. Test the following hypotheses.

We have an two this hypotheses:

1. `Average user ratings of the Xbox One and PC platforms are the same.`
2. `Average user ratings for the Action and Sports genres are different.`


`null` hypothesis states that `mean of the population in question equals A` and `alternative` states that `The mean of the population does not equal A`. So, We will use `st.ttest_ind()` method for testing hypotheses. So we will test for the 1st hypotheses `null` and 2nd as an `alternative`. Before checking hypothese we will apply `Levene`'s test for checking equal variance.

```python
xbox1 = games.query('platform =="XOne"')
pc = games.query('platform == "PC"')
alpha = .05

lev_test1 = st.levene(xbox1['user_score'].dropna(),pc['user_score'].dropna())
print(lev_test1)
result1n = st.ttest_ind(xbox1['user_score'].dropna(),pc['user_score'].dropna(),equal_var=False)
print('p-value of null:',result1n.pvalue)
if (result1n.pvalue < alpha):
    print("We reject the null hypothesis for first hypothesis")
else:
    print("We can't reject the null hypothesis for first hypothesis")


```

```python
action = games.query('genre == "Action"')
sports = games.query('genre == "Sports"')

lev_test2 = st.levene(action['user_score'].dropna(),sports['user_score'].dropna())
print(lev_test2)
result2a = st.ttest_ind(action['user_score'].dropna(),sports['user_score'].dropna(),equal_var=False)
print('p-value of alternative:',result2a.pvalue)
if (result2a.pvalue < alpha):
    print("We reject the null hypothesis for second hypothesis")
else:
    print("We can't reject the null hypothesis for second hypothesis")
```

### Conclusion

We take a significance level as `.05`. And `p-value` stands  probability of any value chosen randomly from sample space will be equal or around `mean`. We set alpha as `.05` to get best subset of the data. And if `p-value` less than `alpha` we will reject hypotheses, else we will accept the hypotheses. And we have set `equal_var` parameter as an `False` in both cases because they differ more than 5%
